**Exercise 7** Gerer um gráfico em setores da mesma variável do gráfico acima e compare os dois gráficos, enfatizando, principalmente, a transmissão da informação para o leitor.

```python
df = pd.DataFrame({'velocidade': [0.1, 17.5, 40, 48, 52, 69, 88], 
                   'tempo de vida': [2, 8, 70, 1.5, 25, 12, 28]}, 
                   index=['caracol', 'porco', 'elefante', 'coelho', 'girafa', 'coiote', 'cavalo'])

df['velocidade'].plot.pie(figsize=(6,6))
```

    O grafico em setores dessa varial já começa a ficar um pouco poluida, um grafico de barra, nesse caso fica mais legivel

**Exercise 8** Explique o resultado obtido quando fazemos **df.plot.bar(y="velocidade", x='tempo de vida')**. E se invertermos  x  e  y ?

```python
df.plot.bar(y="velocidade", x='tempo de vida')
df.plot.bar(x="velocidade", y='tempo de vida')
```

    No primeiro grafico tempo a velocidade em função do tempo de vida, no segundo invertemos isso, obtendo o tempo de vida em função da velocidade
