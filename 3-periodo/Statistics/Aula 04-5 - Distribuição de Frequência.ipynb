### 3- Qual foi a clase que teve a maior e qual foi a que teve a menor porcentagem de mortes?

```python
menor, maior = 100, 0
imenor, imaior = 0, 0
for x in range(1, 4):
    clas = pd.crosstab(train['survived'], train['pclass'], margins=True)[x]
    mortos = (clas[0] / clas[2]) * 100
    
    if menor > mortos:
        menor, imenor = mortos, x
    if maior < mortos:
        maior, imaior = mortos, x 

print(maior, menor)
print(imaior, imenor)
```

** Resposta: **

    -A classe 1 teve a menor porcentagem de mortes com um total de aproximadamente 75,76 %
    -A classe 3 teve a maior porcentagem de mortes com um total de aproximadamente 37,03 %
